```python
from transformers import T5Tokenizer, T5EncoderModel
import torch
import re
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
import seaborn as sns
import umap

from transformers import EsmModel, EsmTokenizer
from sklearn.manifold import TSNE

from sklearn.cluster import KMeans

import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

```

```python
df = pd.read_csv("../NGS-DataAnalysis/Results/CDR3_Enrichment.csv")
class_1 = df[df["R4"] > 42].copy()
class_1["label"] = 1
class_0 = df[df["Enrichment Ratio"] < 1].copy()
class_0["label"] = 0
# Combine class 1 and class 0 into one DataFrame
CDR3Data = pd.concat([class_1, class_0], ignore_index=True)
CDR3Data
```

```python
def get_esm_embeddings_batched(sequences, batch_size=32):
    embeddings = []
    for i in range(0, len(sequences), batch_size):
        batch = sequences[i:i + batch_size]
        inputs = tokenizer(batch, return_tensors="pt", padding=True, truncation=True).to(device)
        with torch.no_grad():
            outputs = model(**inputs)
        batch_embeddings = outputs.last_hidden_state[:, 0, :].cpu().numpy()
        embeddings.append(batch_embeddings)
    return np.vstack(embeddings)

```

```python
# Get embeddings in batches
# Extract sequences and labels
sequences = CDR3Data["CDR3"].tolist()
labels = CDR3Data["label"].tolist()
```

```python
# Initialize ESM2 tokenizer and model
##esm2-650M

tokenizer = EsmTokenizer.from_pretrained("facebook/esm2_t33_650M_UR50D")
model = EsmModel.from_pretrained("facebook/esm2_t33_650M_UR50D")

# Get embeddings
print("Transforming sequences with ESM2_650M parameters...")
embeddings_650M = get_esm_embeddings_batched(sequences, batch_size=16)  

##save embedings
np.save("Supervised_esm2_650M_embeddings.npy", embeddings_650M)
print("Embeddings saved to Supervised_esm2_650M_embeddings.npy")
```

```python
# Initialize ESM2 tokenizer and model
##esm2-3B

tokenizer = EsmTokenizer.from_pretrained("facebook/esm2_t36_3B_UR50D")
model = EsmModel.from_pretrained("facebook/esm2_t36_3B_UR50D")

# Get embeddings
print("Transforming sequences with ESM2_3Bparameters...")
embeddings_3B = get_esm_embeddings_batched(sequences, batch_size=16)  # Adjust batch size as needed
##save embedings
np.save("Supervised_esm2_3B_embeddings.npy", embeddings_3B)
print("Embeddings saved to Supervised_esm2_3B_embeddings.npy")
```

```python


# Create a DataFrame for plotting convenience
plot_data = pd.DataFrame({
    "x_tsne": tsne_results[:, 0],
    "y_tsne": tsne_results[:, 1],
    "x_umap": umap_results[:, 0],
    "y_umap": umap_results[:, 1],
    "CDR3": CDR3Data["CDR3"],
    "Class": CDR3Data["label"],
    "Color": colors
})

# Separate the data into groups for plotting.
# We want:
#   - Orange (lowest zorder; Class 0) in the back,
#   - Green (middle; Class 1) in the middle,
#   - Red and Blue (highlighted) in front.
data_orange   = plot_data[plot_data["Color"] == "orange"]
data_green    = plot_data[plot_data["Color"] == "green"]
data_red_blue = plot_data[plot_data["Color"].isin(["red", "blue"])]

# Create subplots for t-SNE and UMAP
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# -------------------------------
# t-SNE Visualization
# -------------------------------
ax_tsne = axes[0]
# Plot the orange points first (zorder=1)
ax_tsne.scatter(data_orange["x_tsne"], data_orange["y_tsne"],
                color="orange", s=30, alpha=0.7, zorder=1, label="Class 0")
# Then plot the green points (zorder=2)
ax_tsne.scatter(data_green["x_tsne"], data_green["y_tsne"],
                color="green", s=50, alpha=0.9, zorder=2, label="Class 1")
# Finally, plot the red and blue points on top (zorder=3)
ax_tsne.scatter(data_red_blue["x_tsne"], data_red_blue["y_tsne"],
                color=data_red_blue["Color"], s=60, alpha=0.9, zorder=3)

ax_tsne.set_title("t-SNE Visualization of CDR3 Embeddings(ESM2B650M)")
ax_tsne.set_xlabel("t-SNE 1")
ax_tsne.set_ylabel("t-SNE 2")

# -------------------------------
# UMAP Visualization
# -------------------------------
ax_umap = axes[1]
# Plot orange points first (zorder=1)
ax_umap.scatter(data_orange["x_umap"], data_orange["y_umap"],
                color="orange", s=20, alpha=0.7, zorder=1, label="Class 0")
# Plot green points next (zorder=2)
ax_umap.scatter(data_green["x_umap"], data_green["y_umap"],
                color="green", s=50, alpha=0.8, zorder=2, label="Class 1")
# Plot red and blue points on top (zorder=3)
ax_umap.scatter(data_red_blue["x_umap"], data_red_blue["y_umap"],
                color=data_red_blue["Color"], s=60, alpha=0.8, zorder=3)

ax_umap.set_title("UMAP Visualization of CDR3 Embeddings(ESM2B650M)")
ax_umap.set_xlabel("UMAP 1")
ax_umap.set_ylabel("UMAP 2")

# -------------------------------
# Legend
# -------------------------------
handles = [
    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='green', markersize=10, label='Class 1'),
    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='red', markersize=10, label='Highlighted Red'),
    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', markersize=10, label='Highlighted Blue'),
    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='orange', markersize=10, label='Class 0'),
]
fig.legend(handles=handles, loc="lower center", ncol=4, fontsize="medium")

plt.tight_layout()
plt.savefig("CDR3_Embeddings_Visualization_ESM2b650M.png", dpi=600, bbox_inches="tight")
plt.show()
```
